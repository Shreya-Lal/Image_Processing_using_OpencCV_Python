```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("lip.png", cv2.IMREAD_GRAYSCALE)

canny = cv2.Canny(img, 60, 180)

cv2.imshow("Image", img)
cv2.imshow("Canny", canny)

indices = np.where(canny != [0])
coordinates = zip(indices[0], indices[1])

coordinates_list = ""
for coordinate in coordinates:
    x = "'('{}, {}')', ".format(coordinate[1] / 100, -coordinate[0] / 100)
    coordinates_list += x

coordinates_list = "'('{}')'".format(coordinates_list)
coordinates_list = coordinates_list.replace("'('", "{")
coordinates_list = coordinates_list.replace("')'", "}")

print(coordinates_list)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
