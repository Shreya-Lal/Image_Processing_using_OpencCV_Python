# Adaptive Mean Thresholding

```python
# Adaptive_thresh_mean_C

# The threshold value T(x,y) is a mean of the blockSize * blockSize neighborhood of (x,y) minus C
```

```python
# Implementation of Adaptive_thresh_mean_C from scratch

im_gray2 = cv2.imread('thres_imag.jpg', cv2.IMREAD_GRAYSCALE)

h, w = im_gray2.shape[0], im_gray2.shape[1]

avg_kernal = np.ones((11,11), np.uint8)/121
kh, kw = avg_kernal.shape

dark_img = im_gray2.copy()
dark_img[:, :] = 0

for col in range(w-kw+1):
    for row in range(h-kh+1):
        
        dark_img = np.mean(np.sum(np.sum(im_gray2[row:row +kh, col:col+kw] * avg_kernal)))
        
        if(im_gray2[row, col] > (dark_img - 2)):
            im_gray2[row, col] = 255
        else:
            im_gray2[row, col] = 0
            
# plotting result in a grid
#fig, axs = plt.subplots(2,2, figsize = (30,30))

# plt.subplot(1,2,1)
# plt.title("Default Adaptive_thresh_Mean_C Method", fontsize = 20)
# plt.imshow(-----, 'gray')

#plt.subplot(1,2,2)
plt.title("Adaptive_thresh_Mean_C Scratch", fontsize = 20)
plt.imshow(im_gray2, 'gray')
```

```python
# Adaptive_Thresh_Gaussian_C
# The thresh value T(x,y) is a weighted sum(cross-correlation with a Gaussian window) of the blockSize*blockSize neighborhood of (x,y) minus C
# The default sigma (standard deviation) is used for a specified blockSize
```

```python
def gaus_kernal(filter_size, sigma):
    
    # required filter size
    filter_Size = int(filter_size)//2
    
    # creating a blob with that filter size
    ht, wt = np.mgrid[-filter_size:filter_size +1, -filter_size:filter_size+1]
    
    # Using the Gaussian formula for computation
    gaus_constant = 1 / (2.0 * np.pi * sigma**2)
    
    gauss = np.exp(-((ht**2 + wt**2) / (2.0 * sigma**2))) * gaus_constant
    
    return gauss
```

```python
im_gray3 = cv2.imread('thres_imag.jpg', cv2.IMREAD_GRAYSCALE)

h, w = im_gray3.shape[0], im_gray3.shape[1]

kernal = gaus_kernal(11, 2)
kh, kw = kernal.shape

dark_img1 = im_gray3.copy()
dark_img1[:, :] = 0

for row in range(h-kh+1):
    for col in range(w-kw+1):
        dark_img1[row, col]= np.average(np.sum(np.sum(im_gray3[row:row + kh, col:col+kw] * kernal)))
        
        if(im_gray3[row, col] > (dark_img1[row, col] - 2)):
            im_gray3[row, col] = 255
        else:
            im_gray3[row, col] = 0
```

```python
# plotting result in a grid
#fig, axs = plt.subplots(2,2, figsize = (30,30))

# plt.subplot(1,2,1)
# plt.title("Default Adaptive_thresh_Gaussian_C Method", fontsize = 20)
# plt.imshow(-----, 'gray')

#plt.subplot(1,2,2)
plt.title("Adaptive_thresh_Gaussian_C Scratch", fontsize = 20)
plt.imshow(im_gray3, 'gray')
```
