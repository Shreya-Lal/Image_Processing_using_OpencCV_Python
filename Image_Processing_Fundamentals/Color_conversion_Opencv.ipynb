```python
import cv2 as cv
import sys
import os

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## BGR2GRAY

```python
#using opencv inbuilt function
image = cv.imread("red.jpg")
gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)

cv.imshow("Original Image", image)
k= cv.waitKey(0)

cv.imshow("Gray", gray)
plt.imshow(gray, cmap = 'gray')
k= cv.waitKey(0)  # if k = ... is not written -1 is shown in the output after closing window 
```

# BGR2GRAY by changing flag value of imread

```python
# using imread function with flag =0

# Use the second argument or (flag value) zero
# that specifies the image is to be read in grayscale mode
img = cv.imread("red.jpg", 0)
cv.imshow("Original Image", image)
k= cv.waitKey(0)

cv.imshow('Grayscale Image', img)
k= cv.waitKey(0)

if k == ord("s"):
    cv.imwrite("grayscale.jpg", img)

# Window shown waits for any key pressing event
cv.destroyAllWindows()
```
