```python
import cv2 as cv
import sys
import os

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Load Image
img_rgb = cv.imread("bird.jpg")

# splitting the color channels of the image  and then normalizing
r, g, b = img_rgb[:,:,2], img_rgb[:,:,1], img_rgb[:,:,0]
R, G, B = r/ 255.0 , g/255.0, b/255.0

# create a dummy matrix with all values 0
yiq = img_rgb.copy()
yiq[:,:,:] = [0, 0 ,0]

# iterating through the image height and width
for i in range(img_rgb.shape[0]):
    for j in range(img_rgb.shape[1]):
        
        R = img_rgb[i,j,0]
        G = img_rgb[i,j,1]
        B = img_rgb[i,j,2]
        
        yiq[i,j,0] =  0.299 * R + 0.587 * G + 0.114 * B;
        yiq[i,j,1] =  0.596 * R - 0.275 * G - 0.0321 * B;
        yiq[i,j,2] =  0.212 * R - 0.523 * G + 0.311 * B;

        
# plotting the results on grid
fig, axs = plt.subplots(2, 2, figsize = (40, 40))

plt.subplot(1, 2, 1)
plt.title("RGB Image", fontsize = 30)
plt.imshow(img_rgb)

plt.subplot(1, 2, 2)
plt.title("Manual YIQ Image", fontsize = 30)
plt.imshow(cv.cvtColor(yiq, cv.COLOR_BGR2RGB))
```
