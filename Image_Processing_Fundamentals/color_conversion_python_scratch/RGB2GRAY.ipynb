# RGB2GRAY 

```python
import cv2 as cv
import sys
import os

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def rgb2gray(img):
    grayImg=np.zeros(img.shape)
    print(grayImg.shape)
    B = np.array(img[:, :, 0].astype(float))
    G = np.array(img[:, :, 1].astype(float))
    R = np.array(img[:, :, 2].astype(float))

    R = (R *.299)
    G = (G *.587)
    B = (B *.114)
    
    #grayImage = (np.dstack((R,G,B)) ).astype(np.uint8)

    Avg = (R+G+B)
    grayImage = img.copy()

    for i in range(3):
        grayImage[:,:,i] = Avg
           
    return grayImage         
    
image = cv.imread("red.jpg")
grayImage = rgb2gray(image)

cv.imshow("Original Image", image)
key=cv.waitKey(0)

cv.imshow("Grayscale Image",grayImage)
key=cv.waitKey(0)
#plt.show()
```

```python
#Displaying above images 
f, axes = plt.subplots(1,2, figsize = (15,15))
list = [image[:,:,::-1],grayImage]
i = 0
for ax in axes:
    ax.imshow(list[i])
    i+=1
```
